```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
emotions_lifeslice = pd.read_csv('./my_datasets/emotions_lifeslice.csv') 
```

```python
emotions_lifeslice.iloc[70:80]
```

```python
from sklearn.linear_model import LinearRegression
import math

emotions1_train = np.array(emotions_lifeslice['emojis.relaxed'].iloc[0:100]).reshape(-1,1)
emotions2_train = emotions_lifeslice['expressions.lipStretch'].iloc[0:100]
emotions1_test = np.array(emotions_lifeslice['emojis.relaxed'].iloc[300:320]).reshape(-1,1)
emotions2_test = emotions_lifeslice['expressions.lipStretch'].iloc[300:320]

regressor = LinearRegression().fit(emotions1_train, emotions2_train)

print(regressor.score(emotions1_train, emotions2_train))
print(regressor.score(emotions1_test, emotions2_test))

data = np.array([0.01, 0.2, 0.8, 1]).reshape(-1,1)
result = np.array(regressor.predict(data), dtype=float)
plt.plot(data, result, '.')

result
```
