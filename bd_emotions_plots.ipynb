```python
emotions_lifeslice = emotions_lifeslice.iloc[8000:9000]
fig = plt.figure()
axes = fig.add_axes([0.0, 0.0, 1.0, 1.0])
males = emotions_lifeslice[emotions_lifeslice["appearance.gender"] == "Male"]
females = emotions_lifeslice[emotions_lifeslice["appearance.gender"] == "Female"]
unknown = emotions_lifeslice[emotions_lifeslice["appearance.gender"] == "Unknown"]
cols = max(len(males), len(females), len(unknown)) #  количество столбцов

axes.hist(males['appearance.age'], bins=cols, alpha=0.6, label='Мужчины') # добавляем на холст гистограмму распределения возрастов среди мужчин
axes.hist(females['appearance.age'], bins=cols, alpha=0.6, label='Женщины') # добавляем на холст гистограмму распределения возрастов среди женщин
axes.hist(unknown['appearance.age'], bins=cols, alpha=0.6, label='Неопределенные') # добавляем на холст гистограмму распределения возрастов среди неопределенных

axes.legend() # строим легенду
axes.set_xlabel('Возраст', fontsize=18) 
axes.set_ylabel('Количество', fontsize=18)
axes.set_title('Распределение возрастов по полу человека', fontsize=18)
```
