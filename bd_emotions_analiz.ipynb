```python
#10
def change_season(df,date):
    df[date] = df[date].replace(['\d{4}-(12|01|02)-\d{2}'],'winter',regex=True)
    df[date] = df[date].replace(['\d{4}-(03|04|05)-\d{2}'],'spring',regex=True)
    df[date] = df[date].replace(['\d{4}-(06|07|08)-\d{2}'],'summer',regex=True)
    df[date] = df[date].replace(['\d{4}-(09|10|11)-\d{2}'],'autumn',regex=True)
    df.columns = df.columns.str.replace('date','season')
    
emotions_lifeslice_change_date = emotions_columne_byte_lifeslice
change_season(emotions_lifeslice_change_date,'date')
emotions_lifeslice_change_date
```

```python
emotions_lifeslice_change_date_update = emotions_lifeslice_change_date.groupby(['season']).sum()
emotions_lifeslice_change_date_update = emotions_lifeslice_change_date_update.drop(columns=emotions_lifeslice_change_date_update.iloc[:, 0:6])
emotions_lifeslice_change_date_update = emotions_lifeslice_change_date_update.drop(columns=emotions_lifeslice_change_date_update.iloc[:, 8:8])
emotions_lifeslice_change_date_update
```

```python
from pathlib import Path  

datasets_to_save = {
    "emotions_lifeslice": emotions_lifeslice,
    "emotions_columne_byte_lifeslice": emotions_columne_byte_lifeslice,
    "emojis_lifeslice": emojis_lifeslice,
    "expressions_lifeslice": expressions_lifeslice,
    "emotions_lifeslice_sort": emotions_lifeslice_sort,
    "emotions_lifeslice_change_date": emotions_lifeslice_change_date,
    "emotions_lifeslice_change_date_update": emotions_lifeslice_change_date_update
}

path_to_my_datasets = "./my_datasets"

for(file_name, df) in datasets_to_save.items():
    filepath = Path(f'{path_to_my_datasets}/{file_name}.csv')  
    filepath.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(filepath, index=False)
```
